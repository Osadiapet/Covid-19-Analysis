```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('covid_data.csv')
data
```

```python
data.shape
```

```python
#removing all null values from our dataset
data=data.dropna()
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
#total number of cases recorded as at 16/10/2020
cases=data['cases'].sum()
cases
```

```python
#total number of deaths
death=data['deaths'].sum()
death
```

```python
#percentage of death
death_percentage=death/cases*100
death_percentage
```

```python
#survived 
survived=cases-death
survived
```

```python
#percentage of survived patients 
per_survived =survived/cases*100
per_survived
```

```python
#worldwide population as at 2019
world_population=data['popData2019'].sum()
world_population
```

```python
#percestage of people who had the covid worldwide per 100000 persons
covid_persons=cases/world_population*100000
covid_persons
```

```python
continents=data.groupby(['continentExp']).sum()
continents
```

```python
#number of cases recorded by each continent
continents['cases']
```

```python
continents['cases'].plot.bar()
plt.title('Graphical distribution of recorded number of cases for each continents')
plt.xlabel("Continent")
plt.ylabel('Number of cases')
plt.ylim(0,20000000)
plt.show()
```

```python
#number of covid death recorded by each continent
continents['deaths']
```

```python
continents['deaths'].plot.bar()
plt.title('Graphical distribution of recorded number of death cases for each continents')
plt.xlabel("Continent")
plt.ylabel('Number of death cases')
plt.show()
```

```python
continents[['cases', 'deaths']].plot.bar()
plt.title('comparison of recorded cases and number of deaths for each continents')
plt.xlabel("Continent")
plt.ylabel('Number of recorded death cases')
plt.show()
```

```python
continent_covid_percent=continents['cases']/continents['popData2019']*100
continents['continent_covid_percent']=continent_covid_percent
continent_death_percent=continents['deaths']/continents['cases']*100
continents['continent_death_percent']=continent_death_percent
continent_overall_covid_percent=continents['cases']/sum(data['cases'])*100
continents['continent_overall_covid_percent']=continent_overall_covid_percent
continent_overall_death_percent=continents['deaths']/sum(data['deaths'])*100
continents['continent_overall_death_percent']=continent_overall_death_percent
```

```python
continent1=continents[['cases','deaths','continent_death_percent','continent_covid_percent','continent_overall_covid_percent',
                      'continent_overall_death_percent']]
continent1
```

```python
plt.pie(continent1['continent_overall_covid_percent'],labels=('Africa','America','Asia','Europe','Oceania'),autopct='%1.1f%%')
plt.title('percentage of covid cases in the continnts with respect to total recorded cases')
plt.show()
```

```python
plt.pie(continent1['continent_overall_death_percent'],labels=('Africa','America','Asia','Europe','Oceania'), autopct='%1.1f%%' )
plt.title("Percentage of covid death cases with respect to worldwide recorded death cases in the various continents")
plt.show()
```

```python
#death percentage of continents with respect to their population
continent1['continent_death_percent'].plot.bar()
plt.title('Percentage of covid death cases per continent with respect to their total population')
plt.xlabel('Continent')
plt.ylabel('percentage of covid death per recorded cases')
plt.show()
```

```python
#percentage of covid 19 cases of continents with respect to their population 
continent1['continent_covid_percent'].plot.bar()
plt.title('Percentage of covid cases per continent with respect to their total population')
plt.xlabel('Continent')
plt.ylabel('percentage of covid cases per population')
plt.show()
```

```python
continent1[['continent_covid_percent','continent_death_percent']].plot.bar()
plt.title('comparing Percentage of covid death cases with respect to their total cases')
plt.xlabel('Continent')
plt.ylabel('percentage of covid cases per population')
plt.show()
```

```python
death_percent=country['deaths']/country['cases']*100
country['death_percent']=death_percent
covid_percent=country['cases']/country['popData2019']*100
country['covid_percent']=covid_percent
```

```python
#countries that recorded massively with respect to their population
country1['covid_percent'].sort_values(ascending=False)
```

```python
country1['covid_percent'].sort_values(ascending=False).head(20).plot.bar()
plt.title("Graphical representation of countries with high recorded rate of covid with respect to their population ")
plt.xlabel('Country')
plt.ylabel('covid Percentage')
plt.show()
```
